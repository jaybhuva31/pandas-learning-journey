```python
import numpy as np

img = np.random.randint(0, 10, (5,5))

kernel = np.array([
    [1, 0, -1],
    [1, 0, -1],
    [1, 0, -1]
])

result = 0

# Apply kernel on center
for i in range(3):
    for j in range(3):
        result += img[i][j] * kernel[i][j]

print("Result:", result)
```
