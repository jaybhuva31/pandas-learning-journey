```python
# # Day 7 - Data Aggregation & Transformation 🚀

# ## Topics Covered:
# - groupby() for aggregation
# - apply() for transformation
# - merge() for joining data
# - concat() for combining data
```

```python
import pandas as pd
```

```python
data1 = {
    'ID': [1, 2, 3, 4],
    'Name': ['Jay', 'Raj', 'Meet', 'Neel'],
    'Department': ['IT', 'HR', 'IT', 'HR'],
    'Marks': [80, 90, 85, 70]
}

df1 = pd.DataFrame(data1)

data2 = {
    'ID': [1, 2, 3, 4],
    'City': ['Ahmedabad', 'Surat', 'Rajkot', 'Baroda']
}

df2 = pd.DataFrame(data2)

df1
```

```python
df1.groupby('Department')['Marks'].mean()
```

```python
df1.groupby('Department')['Marks'].agg(['mean', 'max', 'min'])
```

```python
df1['Marks'] = df1['Marks'].apply(lambda x: x + 5)
df1
```

```python
merged_df = pd.merge(df1, df2, on='ID')
merged_df
```

```python
df_new = pd.concat([df1, df1])
df_new
```
