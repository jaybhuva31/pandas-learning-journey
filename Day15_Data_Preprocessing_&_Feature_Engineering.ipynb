```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Name': ['Jay', 'Raj', 'Meet', 'Neel', 'Amit'],
    'Age': [20, 21, None, 22, 20],
    'Marks': [80, 90, 85, None, 75],
    'Department': ['IT', 'HR', 'IT', 'HR', 'IT']
}

df = pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Marks'] = df['Marks'].fillna(df['Marks'].mean())
```

```python
df = df.drop('Name', axis=1)
```

```python
df = pd.get_dummies(df, columns=['Department'])
df
```

```python
df['Marks'] = (df['Marks'] - df['Marks'].mean()) / df['Marks'].std()
```

```python
df['Performance'] = df['Marks'] * df['Age']
df
```
