```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Name': ['Jay', 'Raj', 'Meet', 'Neel', 'Amit'],
    'Marks': [80, 90, 85, 70, 300]   # 300 = Outlier 😵
}

df = pd.DataFrame(data)
df
```

```python
Q1 = df['Marks'].quantile(0.25)
Q3 = df['Marks'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print("Lower:", lower)
print("Upper:", upper)
```

```python
outliers = df[(df['Marks'] < lower) | (df['Marks'] > upper)]
outliers
```

```python
df_clean = df[(df['Marks'] >= lower) & (df['Marks'] <= upper)]
df_clean
```

```python
plt.boxplot(df['Marks'])
plt.title("Before Removing Outliers")
plt.show()
```

```python
plt.boxplot(df_clean['Marks'])
plt.title("After Removing Outliers")
plt.show()
```

```python
mean = df['Marks'].mean()
std = df['Marks'].std()

df['z_score'] = (df['Marks'] - mean) / std

df
```

```python
df[df['z_score'] > 3]
```
