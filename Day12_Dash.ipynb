```python
import dash
from dash import dcc, html
import plotly.express as px
import pandas as pd
```

```python
data = {
    'Name': ['Jay', 'Raj', 'Meet', 'Neel'],
    'Marks': [80, 90, 85, 70]
}

df = pd.DataFrame(data)
df
```

```python
fig = px.bar(df, x='Name', y='Marks', title="Student Marks Dashboard")
```

```python
app = dash.Dash(__name__)
```

```python
app.layout = html.Div([
    html.H1("Student Dashboard"),
    dcc.Graph(figure=fig)
])
```

```python
if __name__ == '__main__':
    app.run(debug=True)
```
