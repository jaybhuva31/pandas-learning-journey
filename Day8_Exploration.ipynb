```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
from pandas.plotting import parallel_coordinates
import matplotlib.pyplot as plt
```

```python
data = {
    'Name': ['Jay', 'Raj', 'Meet', 'Neel', 'Amit'],
    'Department': ['IT', 'HR', 'IT', 'HR', 'IT'],
    'Marks': [80, 90, 85, 70, 75],
    'Age': [20, 21, 19, 22, 20]
}

df = pd.DataFrame(data)
df
```

```python
df.sort_values(by='Marks')  # Ascending
```

```python
df.sort_values(by='Marks', ascending=False)  # Descending
```

```python
df[df['Marks'] > 80]
```

```python
df[(df['Department'] == 'IT') & (df['Marks'] > 75)]
```

```python
df['Department'].value_counts()
```

```python
df.corr(numeric_only=True)
```

```python
scatter_matrix(df[['Marks', 'Age']])
plt.show()
```

```python


parallel_coordinates(df[['Marks', 'Age', 'Department']], 'Department')
plt.show()
```
