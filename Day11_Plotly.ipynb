```python
import pandas as pd
import plotly.express as px
```

```python
data = {
    'Name': ['Jay', 'Raj', 'Meet', 'Neel', 'Amit'],
    'Department': ['IT', 'HR', 'IT', 'HR', 'IT'],
    'Marks': [80, 90, 85, 70, 75],
    'Age': [20, 21, 19, 22, 20]
}

df = pd.DataFrame(data)
df
```

```python
fig = px.bar(df, x='Name', y='Marks', title="Student Marks")
fig.show()
```

```python
fig = px.line(df, x='Name', y='Marks', title="Marks Trend")
fig.show()
```

```python
fig = px.pie(df, names='Department', values='Marks', title="Department Distribution")
fig.show()
```

```python
fig = px.scatter(df, x='Age', y='Marks', color='Department',
                 title="Age vs Marks")
fig.show()
```

```python
fig = px.histogram(df, x='Marks', title="Marks Distribution")
fig.show()
```

```python
fig = px.bar(df, x='Name', y='Marks', color='Department',
             title="Student Marks by Department")
fig.show()
```
