```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
G.add_node("Jay")
G.add_node("Raj")
G.add_node("Meet")
```

```python
G.add_edge("Jay", "Raj")
G.add_edge("Jay", "Meet")
```

```python
nx.draw(G, with_labels=True)
plt.show()
```

```python
nx.draw(G, with_labels=True)
plt.show()
```

```python
nx.draw(G, with_labels=True, node_size=2000, font_size=12)
plt.show()
```

```python
G.add_edge("Raj", "Meet")
G.add_edge("Meet", "Neel")

nx.draw(G, with_labels=True)
plt.show()
```

```python
print("Nodes:", G.nodes())
print("Edges:", G.edges())
```
