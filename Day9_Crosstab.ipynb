```python
# # Day 9 - Qualitative vs Quantitative Data 🚀

# ## Topics Covered:
# - Qualitative (Categorical) Data
# - Quantitative (Numeric) Data
# - Two-way Cross Tabulation
# - Crosstab with aggregation
# - Percentage analysis
```

```python
import pandas as pd
```

```python
data = {
    'Name': ['Jay', 'Raj', 'Meet', 'Neel', 'Amit'],
    'Department': ['IT', 'HR', 'IT', 'HR', 'IT'],   # Qualitative
    'Gender': ['Male', 'Male', 'Male', 'Male', 'Male'],  # Qualitative
    'Marks': [80, 90, 85, 70, 75]   # Quantitative
}

df = pd.DataFrame(data)
df
```

```python
pd.crosstab(df['Department'], df['Gender'])
```

```python
pd.crosstab(df['Department'], df['Gender'], values=df['Marks'], aggfunc='mean')
```

```python
pd.crosstab(df['Department'], df['Gender'], margins=True)
```

```python
pd.crosstab(df['Department'], df['Gender'], normalize=True)
```

```python
pd.crosstab(df['Department'], df['Gender']).plot(kind='bar')
```
