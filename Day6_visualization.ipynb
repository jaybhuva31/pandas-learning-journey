```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Name': ['Jay', 'Raj', 'Meet', 'Neel'],
    'Marks': [80, 90, 85, 70]
}

df = pd.DataFrame(data)
df
```

```python
plt.plot(df['Name'], df['Marks'])
plt.title("Marks Line Chart")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
plt.bar(df['Name'], df['Marks'])
plt.title("Marks Bar Chart")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
plt.pie(df['Marks'], labels=df['Name'], autopct='%1.1f%%')
plt.title("Marks Distribution")
plt.show()
```

```python
plt.hist(df['Marks'])
plt.title("Marks Distribution Histogram")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.bar(df['Name'], df['Marks'])
plt.title("Student Marks", fontsize=15)
plt.xlabel("Name")
plt.ylabel("Marks")
plt.grid()
plt.show()
```
